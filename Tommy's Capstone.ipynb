```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues)
nearby_venues.head()
```

```python
filtered_columns = ['venue.name', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head()
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=15)

folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='BEDOKVILLE',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)
```

```python
one_hot = pd.get_dummies(nearby_venues[['name']], prefix="", prefix_sep="")
one_hot['name'] = nearby_venues['name'] 
fixed_columns = [one_hot.columns[-1]] + list(one_hot.columns[:-1])
one_hot = one_hot[fixed_columns]
one_hot.head()
```

```python
one_hot1 = one_hot.groupby(["name"]).mean().reset_index()
one_hot1.head()
```

```python
clusters = 3
clustering = one_hot1.drop(["name"], 1)
kmeans = KMeans(n_clusters=clusters, random_state=0).fit(clustering)
kmeans.labels_[0:10]
one_hot2 = one_hot1.copy()
one_hot2["Cluster"] = kmeans.labels_
one_hot2.head()
```

```python
one_hot2 = one_hot2.join(nearby_venues.set_index("name"), on="name")
one_hot2.head()
```
